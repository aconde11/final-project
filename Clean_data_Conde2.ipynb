```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pylab
import seaborn as sns
sns.set_style("whitegrid")

#import data
path = "appointment.csv"

noshow = pd.read_csv(path)
```

```python
noshow.rename(columns = {'ApointmentData':'AppointmentData',
                         'Alcoolism': 'Alchoholism',
                         'HiperTension': 'Hypertension',
                         'Handcap': 'Handicap'}, inplace = True)

print(noshow.columns)
```

```python
#prob showing up with diff variables
import pandas as pd

def probStatus(noshow, group_by):
    # Create a contingency table
    contingency_table = pd.crosstab(index=noshow[group_by], columns=noshow['status']).reset_index()

    # Calculate the probability of showing up
    contingency_table['probShowUp'] = contingency_table['Show-Up'] / (contingency_table['Show-Up'] + contingency_table['No-Show'])

    # Return the result with the specified columns
    return contingency_table[[group_by, 'probShowUp']]

# Example usage:
# Replace 'your_dataset' and 'your_group_by_column' with your actual dataset and column name
result = probStatus(noshow, 'probShowUp')
print(result)
```

```python
sns.lmplot(data = probStatus(noshow, 'Age'), x = 'Age', y = 'probShowUp', fit_reg = True)
sns.plt.xlim(0, 100)
sns.plt.title('Probability of showing up with respect to Age')
sns.plt.show()

sns.lmplot(data = probStatus(noshow, 'HourOfTheDay'), x = 'HourOfTheDay', 
           y = 'probShowUp', fit_reg = True)
sns.plt.title('Probability of showing up with respect to HourOfTheDay')
sns.plt.show()

sns.lmplot(data = probStatus(noshow, 'AwaitingTime'), x = 'AwaitingTime', 
           y = 'probShowUp', fit_reg = True)
sns.plt.title('Probability of showing up with respect to AwaitingTime')
sns.plt.ylim(0, 1)
sns.plt.show()
```
